```python
import matplotlib.pyplot as plt
import pandas as pd
import csv

# 从CSV文件中读取数据
data = pd.read_csv('average_output.csv')

# 提取所需的数据列
schedule = data['Schedule']
loop1_avg_time = data['Loop 1 Avg Time']
loop2_avg_time = data['Loop 2 Avg Time']

# 创建两个子图，分别绘制 "Loop 1 Avg Time" 和 "Loop 2 Avg Time"
plt.figure(figsize=(24, 12))

# 绘制 "Loop 1 Avg Time" 直方图
plt.subplot(1, 2, 1)
# plt.bar(schedule, loop1_avg_time)
bars = plt.bar(schedule, loop1_avg_time)
plt.title('Loop 1 Avg Time')
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

# 绘制 "Loop 2 Avg Time" 直方图
plt.subplot(1, 2, 2)
#plt.bar(schedule, loop2_avg_time)
bars = plt.bar(schedule, loop2_avg_time)
plt.title('Loop 2 Avg Time')
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')
# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```

```python
data = pd.read_csv('loop1_output.csv')

# 提取数据
threads = data['Num of Thread']
avg_time = data['Avg Time']
speedup = data['Speedup']

# 绘制平均时间的直方图
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(threads, avg_time, color='blue', alpha=0.7)
plt.xlabel('Number of Threads')
plt.ylabel('Average Time')
plt.title('Average Time vs. Number of Threads')

# 绘制速度提升的折线图
plt.subplot(1, 2, 2)
plt.plot(threads, speedup, marker='o', color='green', linestyle='-')
plt.xlabel('Number of Threads')
plt.ylabel('Speedup')
plt.title('Speedup vs. Number of Threads')

# 显示图表
plt.tight_layout()
plt.show()
```
